```python
#Cargamos las librerias
import sklearn
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (16,9)
plt.style.use("fast")

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/drive/MyDrive/Modelos/pokemon.csv")
df.head()
```

```python
df["base_total"]=df["base_total"].astype(float)
df.to_csv("updated_data.csv", index=False)
```

```python
df.dtypes
```

```python
df2 = df[["type1", "type2", "base_total", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed","is_legendary"]]
df2
```

```python
label_encoder = LabelEncoder()
df2["type1"] = label_encoder.fit_transform(df2["type1"])
df2["type2"] = label_encoder.fit_transform(df2["type2"].fillna("None"))
```

```python
caracteristicas = ["type1", "type2", "base_total", "hp", "attack", "defense", "sp_attack", "sp_defense", "speed"]
objetivo = "is_legendary"

X = df2[caracteristicas]
y = df2[objetivo]
```

```python
escala = StandardScaler()
X = escala.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
exactitud = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {exactitud[1]*100:.2f}%")
```
